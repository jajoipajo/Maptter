```python
from pathlib import Path
import re
import lyricsgenius

# --- Config ---
ARTIST_NAME = "Matter"
MAX_SONGS = 500
OUTPUT_DIR = Path("output") / ARTIST_NAME


def load_token_from_env_file(env_path: Path) -> str | None:
    if not env_path.exists():
        return None
    for line in env_path.read_text(encoding="utf-8").splitlines():
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        if line.startswith("GENIUS_ACCESS_TOKEN="):
            return line.split("=", 1)[1].strip().strip('"').strip("'")
    return None


def sanitize_filename(name: str) -> str:
    cleaned = re.sub(r'[<>:\\"/|?*]', "_", name)
    cleaned = cleaned.strip().rstrip(".")
    return cleaned or "untitled"


token = load_token_from_env_file(Path(".env"))
if not token or "PASTE_YOUR_TOKEN_HERE" in token:
    raise ValueError("Missing valid GENIUS_ACCESS_TOKEN in .env")

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

genius = lyricsgenius.Genius(token)
genius.verbose = False
genius.remove_section_headers = True
genius.skip_non_songs = True
genius.excluded_terms = ["(Remix)", "(Live)"]

artist = genius.search_artist(ARTIST_NAME, max_songs=MAX_SONGS, sort="title")
if artist is None:
    raise ValueError(f"Artist not found: {ARTIST_NAME}")

name_counts = {}
saved = 0
for song in artist.songs:
    base = sanitize_filename(song.title)
    count = name_counts.get(base, 0) + 1
    name_counts[base] = count
    filename = f"{base}.txt" if count == 1 else f"{base} ({count}).txt"
    path = OUTPUT_DIR / filename

    lyrics = song.lyrics or ""
    path.write_text(lyrics, encoding="utf-8")
    saved += 1

print(f"Artist: {artist.name}")
print(f"Songs fetched: {len(artist.songs)}")
print(f"Saved txt files: {saved}")
print(f"Folder: {OUTPUT_DIR.resolve()}")
```

```python
from pathlib import Path
import csv
import re
import unicodedata
import geonamescache

LYRICS_DIR = Path("output") / "Matter"
ANNOTATED_DIR = LYRICS_DIR / "annotated_toponyms"
REPORT_PATH = LYRICS_DIR / "toponym_report.csv"


def normalize(text: str) -> str:
    text = unicodedata.normalize("NFKD", text)
    text = "".join(c for c in text if not unicodedata.combining(c))
    return text.casefold()


def collect_place_names() -> set[str]:
    gc = geonamescache.GeonamesCache()
    names: set[str] = set()
    for country in gc.get_countries().values():
        names.add(normalize(country["name"]))
    for city in gc.get_cities().values():
        names.add(normalize(city["name"]))
    return {n for n in names if len(n) >= 3}


PLACE_NAMES = collect_place_names()
token_pattern = re.compile(r"[A-Za-zÀ-ÖØ-öø-ÿČčŠšŽžĆćĐđ]{3,}")


def find_toponyms(text: str):
    matches = []
    seen_spans = set()
    # 1-word and 2-word candidates to catch entries like New York
    words = [(m.group(0), m.start(), m.end()) for m in token_pattern.finditer(text)]
    for i, (w1, s1, e1) in enumerate(words):
        one = normalize(w1)
        if one in PLACE_NAMES:
            key = (s1, e1)
            if key not in seen_spans:
                matches.append((s1, e1, w1))
                seen_spans.add(key)
        if i + 1 < len(words):
            w2, s2, e2 = words[i + 1]
            if s2 == e1 + 1:  # contiguous with one separator char
                two = normalize(f"{w1} {w2}")
                if two in PLACE_NAMES:
                    key = (s1, e2)
                    if key not in seen_spans:
                        matches.append((s1, e2, f"{w1} {w2}"))
                        seen_spans.add(key)
    return sorted(matches, key=lambda x: x[0])


def annotate_text(text: str, spans):
    if not spans:
        return text
    parts = []
    cursor = 0
    for start, end, surface in spans:
        if start < cursor:
            continue
        parts.append(text[cursor:start])
        parts.append(f"[[TOPONYM:{surface}]]")
        cursor = end
    parts.append(text[cursor:])
    return "".join(parts)


ANNOTATED_DIR.mkdir(parents=True, exist_ok=True)
rows = []
files_processed = 0
files_with_toponyms = 0
total_mentions = 0

for txt_path in sorted(LYRICS_DIR.glob("*.txt")):
    files_processed += 1
    text = txt_path.read_text(encoding="utf-8", errors="ignore")
    spans = find_toponyms(text)
    unique_in_file = sorted({normalize(s[2]) for s in spans})
    if spans:
        files_with_toponyms += 1
        total_mentions += len(spans)
    for norm_name in unique_in_file:
        rows.append({"word": norm_name, "file": txt_path.name, "mentions_in_file": sum(1 for s in spans if normalize(s[2]) == norm_name)})
    annotated = annotate_text(text, spans)
    (ANNOTATED_DIR / txt_path.name).write_text(annotated, encoding="utf-8")

rows.sort(key=lambda r: (r["word"], r["file"]))
with REPORT_PATH.open("w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["word", "file", "mentions_in_file"])
    writer.writeheader()
    writer.writerows(rows)

print(f"Processed files: {files_processed}")
print(f"Files with toponyms: {files_with_toponyms}")
print(f"Total toponym mentions: {total_mentions}")
print(f"Annotated files dir: {ANNOTATED_DIR.resolve()}")
print(f"Report CSV: {REPORT_PATH.resolve()}")
```

```python
from pathlib import Path
import csv
import re
import unicodedata
from collections import Counter

LYRICS_DIR = Path("output") / "Matter"
ANNOTATED_DIR = LYRICS_DIR / "annotated_toponyms_sl"
REPORT_PATH = LYRICS_DIR / "toponym_report_sl.csv"

# Strogo kuriran seznam toponimov (slovenske in svetovne oblike v slovenščini)
TOPONYM_BASES = {
    "slovenija", "ljubljana", "maribor", "celje", "kranj", "koper", "novo mesto",
    "nova gorica", "murska sobota", "ptuj", "velenje", "trbovlje", "kamnik", "domzale", "duplica",
    "milano", "rim", "pariz", "london", "berlin", "dunaj", "benetke", "madrid", "barcelona",
    "tokio", "seul", "peking", "bejrut", "new york", "los angeles", "miami", "chicago",
    "kalifornija", "arizona", "texas", "sumatra", "mehika", "amerika", "francija", "italija",
    "nemcija", "avstrija", "hrvaska", "bosna", "srbija", "babilon"
}


def normalize(text: str) -> str:
    text = unicodedata.normalize("NFKD", text)
    text = "".join(c for c in text if not unicodedata.combining(c))
    text = text.replace("đ", "d")
    return text.casefold()


def inflect_word_sl(word: str) -> set[str]:
    w = normalize(word)
    forms = {w}
    if len(w) < 3:
        return forms

    if w.endswith("ija"):
        stem = w[:-1]  # slovenij-
        forms.update({stem + "a", stem + "e", stem + "i", stem + "o", stem + "i", stem + "ama"})
    elif w.endswith("a"):
        stem = w[:-1]
        forms.update({stem + "e", stem + "i", stem + "o", stem + "u", stem + "ama", stem + "ah"})
    elif w.endswith("o"):
        stem = w[:-1]
        forms.update({w, stem + "a", stem + "u", stem + "om", stem + "o", stem + "ju", stem + "jem", stem + "jo"})
    else:
        forms.update({w + "a", w + "u", w + "om", w + "em", w + "i", w + "ih"})

    return forms


def generate_variants(base: str) -> set[str]:
    parts = [p for p in normalize(base).split() if p]
    if not parts:
        return set()
    if len(parts) == 1:
        return inflect_word_sl(parts[0])

    # Pri vecbesednih toponimih sklanjamo zadnjo besedo.
    prefix = " ".join(parts[:-1])
    last_forms = inflect_word_sl(parts[-1])
    return {prefix + " " + lf for lf in last_forms}


variant_to_base = {}
for base in TOPONYM_BASES:
    for v in generate_variants(base):
        variant_to_base.setdefault(v, base)

# Dodatni ročni aliasi, ki se pogosto pojavijo v rap besedilih.
variant_to_base.update({
    "tokijo": "tokio",
    "tokia": "tokio",
    "kamnk": "kamnik",
    "duplici": "duplica",
    "mehike": "mehika",
    "berlina": "berlin",
})

word_re = re.compile(r"[A-Za-zÀ-ÖØ-öø-ÿČčŠšŽžĆćĐđ]{3,}")


def find_toponyms(text: str):
    words = [(m.group(0), m.start(), m.end()) for m in word_re.finditer(text)]
    matches = []
    i = 0
    while i < len(words):
        w1, s1, e1 = words[i]
        chosen = None

        # Poskusi 2-besedni toponim najprej (npr. New York).
        if i + 1 < len(words):
            w2, s2, e2 = words[i + 1]
            if s2 <= e1 + 2:
                two_norm = normalize(w1 + " " + w2)
                if two_norm in variant_to_base:
                    chosen = (s1, e2, text[s1:e2], variant_to_base[two_norm], two_norm)

        if chosen is None:
            one_norm = normalize(w1)
            if one_norm in variant_to_base:
                chosen = (s1, e1, w1, variant_to_base[one_norm], one_norm)

        if chosen is not None:
            matches.append(chosen)
            i += 2 if chosen[1] > e1 else 1
        else:
            i += 1

    return matches


def annotate_text(text: str, spans):
    if not spans:
        return text
    out = []
    cur = 0
    for s, e, surface, canonical, _ in spans:
        if s < cur:
            continue
        out.append(text[cur:s])
        out.append(f"[[TOPONYM:{surface}|{canonical}]]")
        cur = e
    out.append(text[cur:])
    return "".join(out)


ANNOTATED_DIR.mkdir(parents=True, exist_ok=True)
rows = []
files_processed = 0
files_with_toponyms = 0
total_mentions = 0
toponym_counter = Counter()

for txt_path in sorted(LYRICS_DIR.glob("*.txt")):
    files_processed += 1
    text = txt_path.read_text(encoding="utf-8", errors="ignore")
    spans = find_toponyms(text)

    if spans:
        files_with_toponyms += 1
        total_mentions += len(spans)

    per_file = Counter((canonical, matched_norm) for _, _, _, canonical, matched_norm in spans)
    for (canonical, matched_norm), n in per_file.items():
        rows.append({
            "canonical_toponym": canonical,
            "matched_form": matched_norm,
            "file": txt_path.name,
            "mentions_in_file": n,
        })
        toponym_counter[canonical] += n

    annotated_text = annotate_text(text, spans)
    (ANNOTATED_DIR / txt_path.name).write_text(annotated_text, encoding="utf-8")

rows.sort(key=lambda r: (r["canonical_toponym"], r["file"]))
with REPORT_PATH.open("w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["canonical_toponym", "matched_form", "file", "mentions_in_file"])
    writer.writeheader()
    writer.writerows(rows)

print(f"Processed files: {files_processed}")
print(f"Files with toponyms: {files_with_toponyms}")
print(f"Total toponym mentions: {total_mentions}")
print("Top toponyms:")
for name, n in toponym_counter.most_common(20):
    print(f"  {name}: {n}")
print(f"Annotated files dir: {ANNOTATED_DIR.resolve()}")
print(f"Report CSV: {REPORT_PATH.resolve()}")
```

```python
from pathlib import Path
import csv
from collections import defaultdict
import folium
from folium.plugins import MarkerCluster

REPORT_IN = Path("output") / "Matter" / "toponym_report_sl.csv"
TABLE_OUT = Path("output") / "Matter" / "toponym_table_with_coords.csv"
MAP_OUT = Path("output") / "Matter" / "toponym_map.html"

# Kurirane koordinate za toponime v korpusu.
COORDS = {
    "ljubljana": (46.0569, 14.5058),
    "kamnik": (46.2259, 14.6121),
    "duplica": (46.2065, 14.6008),
    "koper": (45.5481, 13.7302),
    "milano": (45.4642, 9.19),
    "rim": (41.9028, 12.4964),
    "berlin": (52.52, 13.405),
    "bejrut": (33.8938, 35.5018),
    "tokio": (35.6762, 139.6503),
    "sumatra": (0.5897, 101.3431),
    "mehika": (23.6345, -102.5528),
    "arizona": (34.0489, -111.0937),
    "texas": (31.9686, -99.9018),
    "chicago": (41.8781, -87.6298),
    "amerika": (39.8283, -98.5795),
    "bosna": (43.9159, 17.6791),
    "italija": (41.8719, 12.5674),
    "babilon": (32.5422, 44.4200),
}

aggregated = defaultdict(int)
with REPORT_IN.open(encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for r in reader:
        toponym = r["canonical_toponym"].strip().lower()
        if toponym not in COORDS:
            continue
        song = r["file"]
        aggregated[(toponym, song)] += int(r["mentions_in_file"])

rows = []
for (toponym, song), count in sorted(aggregated.items()):
    lat, lon = COORDS[toponym]
    rows.append({
        "beseda": toponym,
        "skladba": song,
        "pojavnost": count,
        "lat": lat,
        "lon": lon,
    })

with TABLE_OUT.open("w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["beseda", "skladba", "pojavnost", "lat", "lon"])
    writer.writeheader()
    writer.writerows(rows)

# Agregacija za popup: ena tocka na toponim/lokacijo, notri vse skladbe + pojavnosti.
by_toponym = defaultdict(list)
for row in rows:
    by_toponym[row["beseda"]].append(row)

center = [46.15, 14.99]
m = folium.Map(location=center, zoom_start=5, tiles="CartoDB positron")
cluster = MarkerCluster().add_to(m)

for toponym, items in sorted(by_toponym.items()):
    lat = items[0]["lat"]
    lon = items[0]["lon"]
    total_mentions = sum(i["pojavnost"] for i in items)

    song_lines = "".join(
        f"<li>{i['skladba']}: <b>{i['pojavnost']}</b></li>" for i in sorted(items, key=lambda x: (-x['pojavnost'], x['skladba']))
    )
    html = (
        f"<div style='min-width:280px'>"
        f"<h4 style='margin:0 0 8px 0'>{toponym}</h4>"
        f"<div><b>Koordinata:</b> {lat:.4f}, {lon:.4f}</div>"
        f"<div><b>Skupna pojavnost:</b> {total_mentions}</div>"
        f"<div><b>Skladbe:</b></div><ul>{song_lines}</ul>"
        f"</div>"
    )

    folium.CircleMarker(
        location=[lat, lon],
        radius=min(16, 5 + total_mentions ** 0.5),
        color="#b22222",
        fill=True,
        fill_opacity=0.7,
        popup=folium.Popup(html, max_width=420),
        tooltip=f"{toponym} ({total_mentions})",
    ).add_to(cluster)

m.save(str(MAP_OUT))

print(f"Rows in table: {len(rows)}")
print(f"Table: {TABLE_OUT.resolve()}")
print(f"Map: {MAP_OUT.resolve()}")
```
